```python
df.drop_duplicates(inplace=True)

# Fill missing country with 'Unknown'
df['country'].fillna('Unknown', inplace=True)

# Fill missing director with 'Unknown'
df['director'].fillna('Unknown', inplace=True)

# Fill missing rating with 'Not Rated'
df['rating'].fillna('Not Rated', inplace=True)
```

```python
print(" EDA Insights Summary:")
print(f"• Total Records: {df.shape[0]}")
print(f"• Total Movies: {df[df['type']=='Movie'].shape[0]}")
print(f"• Total TV Shows: {df[df['type']=='TV Show'].shape[0]}")
print("• Most Common Rating:", df['rating'].mode()[0])
print("• Most Common Country:", df['country'].mode()[0])
print("• Year with Most Releases:", df['release_year'].value_counts().idxmax())

```

```python
# For TV Shows, extract number of seasons
df_tv = df[df['type']=='TV Show'].copy()
df_tv['num_seasons'] = df_tv['duration'].str.replace(' Season','').str.replace('s','').astype(float)

# Merge back to main df
df.loc[df['type']=='TV Show', 'num_seasons'] = df_tv['num_seasons']

# Check
df[['title','type','duration','num_seasons']].head(10)
```

```python
# Movie duration bins
df_movies = df[df['type'] == 'Movie'].copy()
df_movies['duration_num'] = df_movies['duration'].str.extract('(\d+)').astype(float)
bins = [0, 60, 120, 180, 500]
labels = ['Short', 'Medium', 'Long', 'Very Long']
df_movies['duration_category'] = pd.cut(df_movies['duration_num'], bins=bins, labels=labels)

# Check distribution
display(df_movies['duration_category'].value_counts())
```

```python
def rating_group(rating):
    if rating in ['TV-MA', 'R', 'NC-17']:
        return 'Adult'
    elif rating in ['PG-13', 'TV-14']:
        return 'Teen'
    elif rating in ['G', 'TV-G', 'TV-Y', 'TV-Y7']:
        return 'Kids'
    else:
        return 'Other'

df['rating_group'] = df['rating'].apply(rating_group)

df[['title', 'rating', 'rating_group']].head(10)

```

```python
sns.countplot(data=df, x='rating_group', hue='type', palette='muted')
plt.title("Distribution of Audience Rating Groups by Type")
plt.xlabel("Rating Group")
plt.ylabel("Count")
plt.show()
```
