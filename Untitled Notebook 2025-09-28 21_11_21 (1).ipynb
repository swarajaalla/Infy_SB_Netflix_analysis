COUNTRY AS 1 AND UNKNOWN AS 0


```python
import pandas as pd

df = pd.read_csv("netflix_cleaned.csv")

# Fill missing countries
df["country"] = df["country"].fillna("Unknown").str.strip()

# Take the first country if multiple
df["main_country"] = df["country"].str.split(",").str[0].str.strip()

# Create binary code: 1 if country is known, 0 if Unknown
df["country_code"] = df["main_country"].apply(lambda x: 0 if x.lower() == "unknown" else 1)

# Quick peek
df[["country", "main_country", "country_code"]].head(20)
```

Creating Numeric Flag for Movie Type movie as 1 and tvshow as 0

```python
# Convert type into numeric category
df["is_movie"] = df["type"].apply(lambda x: 1 if x.strip().lower() == "movie" else 0)

# Quick check
df[["type", "is_movie"]].head(10)

```

Identifying TV Shows Using Duration and Unit Extraction and min as 0 and season as 1


```python
import pandas as pd

# Read cleaned dataset
df = pd.read_csv("netflix_cleaned.csv")

# Fill missing duration and extract unit
df["duration"] = df["duration"].fillna("0 min")
df["duration_unit"] = df["duration"].str.extract(r"\d+\s*(\w+)")[0]

# Create binary column: 1 if Seasons, 0 if min
df["is_tv_show"] = df["duration_unit"].str.lower().apply(lambda x: 1 if x == "seasons" else 0)

# Quick peek
df[["duration", "duration_unit", "is_tv_show"]].head(10)
```

Encoding and Normalizing Rating Frequencies calculate norm freq for rating

```python
import pandas as pd

# Read cleaned dataset
df = pd.read_csv("netflix_cleaned.csv")

# Example: frequency encode the 'rating' column
freq = df["rating"].value_counts()       # count of each category
df["rating_freq"] = df["rating"].map(freq)  # map counts back to the dataframe

# Optional: normalize to get proportion instead of count
df["rating_freq_norm"] = df["rating"].map(freq / len(df))

# Quick check
df[["rating", "rating_freq", "rating_freq_norm"]].head(10)
```

Country Label Encoding and Sample Output 

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Read cleaned dataset
df = pd.read_csv("netflix_cleaned.csv")

# Fill missing values in main_country
df["main_country"] = df["country"].fillna("Unknown").str.split(",").str[0].str.strip()

# Apply Label Encoding
le_country = LabelEncoder()
df["main_country_label"] = le_country.fit_transform(df["main_country"])

# Quick check
df[["main_country", "main_country_label"]].head(10)
```

Director Name Cleaning and Movie Type Numeric Encoding and movie director as 1 and other as 0

```python
import pandas as pd

# Read cleaned dataset
df = pd.read_csv("netflix_cleaned.csv")

# Fill missing director names
df["director"] = df["director"].fillna("Unknown Director").str.strip()

# Create numeric director column based on type
df["director_code"] = df.apply(lambda row: 1 if str(row["type"]).strip().lower() == "movie" else 0, axis=1)

# Quick peek
df[["title", "type", "director", "director_code"]].head(10)
```
